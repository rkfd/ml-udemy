```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
```

```python
# import dataset
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
# train decision tree regression model on whole dataset
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x, y)
```

```python
# predict new result
regressor.predict([[6.5]])
```

```python
# visualize the decision tree regression results (high resolution and smoother curve)
x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape((len(x_grid), 1))

plt.scatter(x, y, color='red')
plt.plot(x_grid, regressor.predict(x_grid), color='blue')
plt.title('Truth or Bluff (Decision Tree Regression) [High Res and Smooth]')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
